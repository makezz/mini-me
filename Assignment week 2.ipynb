```python
#Question 1
def age_range():
    age = int(input("Please enter your age: "))
    if age <= 12:
        print("Your age is %d and you are a Toddler"%(age))
    elif age <= 19:
        print("Your age is %d and you are a Teenager"%(age))
    elif age <= 35:
        print("Your age is %d and you are a Youth"%(age))
    elif age <= 45:
        print("Your age is %d and you are an Adult"%(age))
    elif age >= 46:
        print("Your age is %d and you are an Elder"%(age))
    else:
        print("Invalid input")
        return age
    
age_range()
```
