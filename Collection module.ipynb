```python
from collections import OrderedDict
```

```python
ordered_dict = OrderedDict()
ordered_dict['my age'] = 26
ordered_dict['my bf'] = 36
print(ordered_dict)
```

```python
from collections import defaultdict
d = defaultdict(float)
d['a'] = 1
d['b'] = 2
print(d)
```

```python
from collections import deque 
d = deque()
d.append(1)
d.append(2)
d.appendleft(3)
print(d)
```

```python
d.popleft()
```

```python
print(d)
```

```python
d.extendleft([4,5,6])
```

```python
print(d)
```

```python
class Employee:
    def __init__(self, first, last, pay):
       self.first = first
       self.last = last
       self.pay = pay
       self.email = first + "." + last + "@company.com"
        
    def fullname(self):
        return '%s %s'%(self.first, self.last)
    def apply_raise(self):
        self.pay = int(self.pay * 1.04)
emp_1 = Employee('Christine', 'Okungbowa', 50000)
emp_2 = Employee('Test', 'User', 60000)
print(emp_1.email)
print(emp_2.email)
print(emp_2.fullname)
```
